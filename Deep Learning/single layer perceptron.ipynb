```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv("data_sets.csv")
df.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="score",data=df,hue="placed")
plt.show()
```

```python
X=df.iloc[:,:-1]
y=df["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr=Perceptron(alpha=0.1)
pr.fit(X_train,y_train)
```

```python
pr.score(X_train,y_train)*100,pr.score(X_test,y_test)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(X.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```
